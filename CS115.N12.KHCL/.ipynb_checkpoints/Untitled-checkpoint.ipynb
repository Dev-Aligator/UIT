```python
# Palmer Penguins dataset https://github.com/mcnakhaee/palmerpenguins
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from palmerpenguins import load_penguins
```

```python
penguins = load_penguins()
penguins

```

```python
palette = {'Adelie': '#FF8C00', 'Gentoo': '#159090', 'Chinstrap': '#A034F0'}
```

```python
g = sns.lmplot(x="flipper_length_mm",
               y="body_mass_g",
               hue="species",
               height=7,
               data=penguins,
               palette= palette)
g.set_xlabels('Flipper Length')
g.set_ylabels('Body Mass')
```

```python
t = sns.displot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species')
```

```python
sns.displot(data=penguins, x='bill_length_mm', y='bill_depth_mm',  hue='species', palette=palette ,binwidth=(15,20))
```

```python
e = sns.jointplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="species",
    kind="kde",
)
```

```python
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm' ,palette = palette, hue="species")
```

```python
sns.displot(data=penguins, x='bill_length_mm', hue='species', palette=palette)
```

```python
sns.displot(data=penguins, x='bill_length_mm', hue='species', palette=palette, kind='kde', fill=True)
```
